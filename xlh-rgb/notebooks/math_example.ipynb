```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3, 3, 21, endpoint=True)
x
```

## Lineare Funktion 1. Grades
$y_l=f(x)=a x + b$

```python
a = 2.0 
b = 1.0

# Funktionsdefinition
yl = a * x + b
yl  # Ausgabe der berechneten Werte
```

## Quadratische Funktion
$y_q=f(x)=c x^2 + d$

```python
c = 2.0 
d = 1.0

# Funktionsdefinition
yq = a * x**2 + d
yq  # Ausgabe der berechneten Werte
```

```python
plt.plot(x, yl, label='y_l=f(x)=ax+b')
plt.plot(x, yq, label='$y_q=f(x)=cx^2$+d')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Funktionen')
pass  # unterdr√ºckt Printausgabe
```
