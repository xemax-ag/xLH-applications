```python
from opcua import Client, ua
import pandas as pd
import webbrowser
```

```python
# IP Adresse der SPS
ip_xlh_base = '192.168.1.151'
```

```python
%%time
# auslesen der Daten aus der SPS via OPC-UA
port_plc = 4840
opcua_client = Client(f'opc.tcp://{ip_xlh_base}:{port_plc}')
opcua_client.connect()
root = opcua_client.get_root_node()
plc_name = "4:CODESYS Control for Raspberry Pi 64 SL"
node_root_base = ['0:Objects', '2:DeviceSet', plc_name,
                  '3:Resources', '4:app', '3:Programs', '4:MAIN', '4:data_logger']
afTime_node = root.get_child(node_root_base + ['4:afTime'])
afChannel_00_node = root.get_child(node_root_base + ['4:afChannel_00'])
afChannel_01_node = root.get_child(node_root_base + ['4:afChannel_01'])
afChannel_02_node = root.get_child(node_root_base + ['4:afChannel_02'])
afChannel_03_node = root.get_child(node_root_base + ['4:afChannel_03'])
afChannel_04_node = root.get_child(node_root_base + ['4:afChannel_04'])
afChannel_05_node = root.get_child(node_root_base + ['4:afChannel_05'])
afChannel_06_node = root.get_child(node_root_base + ['4:afChannel_06'])
afChannel_07_node = root.get_child(node_root_base + ['4:afChannel_07'])
afChannel_08_node = root.get_child(node_root_base + ['4:afChannel_08'])
afChannel_09_node = root.get_child(node_root_base + ['4:afChannel_09'])
afChannel_10_node = root.get_child(node_root_base + ['4:afChannel_10'])
afChannel_11_node = root.get_child(node_root_base + ['4:afChannel_11'])
afChannel_12_node = root.get_child(node_root_base + ['4:afChannel_12'])
afChannel_13_node = root.get_child(node_root_base + ['4:afChannel_13'])
afChannel_14_node = root.get_child(node_root_base + ['4:afChannel_14'])
afChannel_15_node = root.get_child(node_root_base + ['4:afChannel_15'])

afTime = afTime_node.get_value()
afChannel_00 = afChannel_00_node.get_value()
afChannel_01 = afChannel_01_node.get_value()
afChannel_02 = afChannel_02_node.get_value()
afChannel_03 = afChannel_03_node.get_value()
afChannel_04 = afChannel_04_node.get_value()
afChannel_05 = afChannel_05_node.get_value()
afChannel_06 = afChannel_06_node.get_value()
afChannel_07 = afChannel_07_node.get_value()
afChannel_08 = afChannel_08_node.get_value()
afChannel_09 = afChannel_09_node.get_value()
afChannel_10 = afChannel_10_node.get_value()
afChannel_11 = afChannel_11_node.get_value()
afChannel_12 = afChannel_12_node.get_value()
afChannel_13 = afChannel_13_node.get_value()
afChannel_14 = afChannel_14_node.get_value()
afChannel_15 = afChannel_15_node.get_value()

opcua_client.disconnect()
```

```python
# Konvertierung der Rohdaten in ein Pandas Data-Frame (vergleichbar zu Excel)
df = pd.DataFrame({
    't [s]': afTime,
    'relais': afChannel_00,
    'U [V]': afChannel_01,
    'I [A]': afChannel_02,
    'Pel [W]': afChannel_03,
    'Eel [J]': afChannel_04,
    'T1 [°C]': afChannel_05,
    'T2 [°C]': afChannel_06,
    'T3 [°C]': afChannel_07,
    'Tmw [°C]': afChannel_08,    
})
print(df.shape)
df.head(5)
```

```python
# Speicherung der Rohdaten für die weitere Bearbeitung
df.to_pickle('data/measurement_1.pkl')
```
